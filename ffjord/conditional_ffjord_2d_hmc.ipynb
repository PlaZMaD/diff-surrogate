## Evalutation

```python
device = torch.device('cuda:3')
class HybridMC(nn.Module):
    def __init__(self, model, dim_q, L, epsilon, q=None, covariance='unit', t=1.):
        super(HybridMC, self).__init__()
        self._model = model
        self._epsilon = epsilon
        if q is not None:
            self._q = torch.tensor(q).float()
        else:
            self._q = torch.zeros(dim_q, requires_grad=True)
        self._p = torch.zeros_like(self._q)
        self._N = len(self._q)
        self._L = L
        self._t = t
        if covariance == 'unit':
            self._mean_p = torch.zeros(dim_q)
            self._cov_p = torch.eye(dim_q)
        
    def calc_grad_q(self):
        q = self._q.clone().detach()
        q.requires_grad = True
        loss = self._model(q) / self._t
        loss.backward()
        return q.grad
        
    def step(self, t=1.):
        self._t = t
        self._p = dist.MultivariateNormal(self._mean_p, covariance_matrix=self._cov_p).sample(N)
        q_old = self._q.clone().detach()
        p_old = self._p.clone().detach()
        # leap frogs steps
        for i in range(self._L):
            self._p = self._p - (self._epsilon / 2) * self.calc_grad_q()
            self._q = self._q + self._epsilon * self._p
            self._p = self._p - (self._epsilon / 2) * self.calc_grad_q()
            
        # metropolis acceptance step
        with torch.no_grad():
            H_end = (self._model(self._q) / self._t + self._p.pow(2).sum()).item()
            H_start = (self._model(q_old) / self._t + p_old.pow(2).sum()).item()
        
        acc_prob = min(1, np.exp(H_start - H_end))
        if not np.random.binomial(1, acc_prob):
            self._q = q_old.clone().detach()
        return self._q
    
class IntermediateModel(nn.Module):
    def __init__(self, model, partial_condition, dim_in, device):
        super(IntermediateModel, self).__init__()
        self._model = model
        self._dim_in = dim_in
        self._partial_condition = partial_condition
        self.device = device
    
    def forward(self, condition: torch.Tensor):
        conditions = torch.cat([
            condition.repeat(len(self._partial_condition), 1).float().to(self.device), 
            self._partial_condition
        ], dim=1)
        
        samples = self._model(torch.randn(len(conditions), self._dim_in).to(self.device), condition=conditions)
        loss = OptLoss.SigmoidLoss(samples, 5, 10).mean()
        return loss
```

## Parallel HMC

```python
class IntermediateModelParallel(nn.Module):
    def __init__(self, model, partial_condition, dim_in, device):
        super(IntermediateModelParallel, self).__init__()
        self._model = model
        self._dim_in = dim_in
        self._partial_condition = partial_condition
        self.device = device
    
    def forward(self, condition: torch.Tensor):
        n_different_conditions = condition.shape[0]
        conditions = []
        for i in range(n_different_conditions):
            conditions.append(torch.cat([
                condition[i].repeat(len(self._partial_condition), 1).float().to(self.device), 
                self._partial_condition
            ], dim=1))
        conditions = torch.cat(conditions, dim=0)
        
        samples = self._model(torch.randn(len(conditions), self._dim_in).to(self.device), condition=conditions)
        loss = OptLoss.SigmoidLoss(samples, 5, 10).reshape(n_different_conditions, -1).mean(dim=1).sum()
        return loss
```

```python
from torchdiffeq import odeint
class HamiltonianODE(nn.Module):
    def __init__(self, model, dim_q):
        super(HamiltonianODE, self).__init__()
        self._model = model
        self._dim_q = dim_q
        self.nef = 0
        self._temp = 1e-2
        
        
    def calc_grad_q(self, q):
        q = q.clone().detach()
        q.requires_grad = True
        loss = self._model(q) / self._temp
        loss.backward()
        return q.grad
    
    def forward(self, t, x):
        self.nef += 1
        q, p = x[:, :self._dim_q], x[:, self._dim_q:]
        dq = p
        dp = - self.calc_grad_q(q)
        return torch.cat([dq, dp], dim=1)
```
